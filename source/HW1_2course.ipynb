## Аналитическая часть
Теперь необходимо проанализировать полученные результаты:

Составим функцию для анализа:

````python
def analyze(file_path: str, text: str, output: bool=True):
    lengths = []
    total_len, max_len = 0, 0
    num, length, score = 0, 0, 0
    max_sequence, curr_sequence = '', ''

    with open(file_path, 'r') as fin:
        for line in fin:

            if line.startswith('>'):
                if num:
                    lengths.append(length)
                num += 1

                if length >= max_len:
                    max_len, max_sequence = length, curr_sequence

                length, curr_sequence = 0, ''

            else:
                curr_sequence += line.strip()
                length += len(line.strip())
                total_len += len(line.strip())

    lengths.sort(reverse=True)

    for i, length in enumerate(lengths, 1):
        if sum(lengths[:i]) >= total_len / 2:
            if output:
                print(f'Анализируем {text}:',
                      f'Общее количество: {num}',
                      f'Общая длина: {total_len}',
                      f'Длина самого длинного: {max_len}',
                      f'N50: {length}',
                      sep='\n')
            break
    return max_sequence
````

Результат для контигов:

````python
max_cont = analyze('Poil_contig.fa', 'контиги')
````

Результат для скафолдов:

````python
max_scaf = analyze('Poil_scaffold.fa', 'cкаффолды')
````

Сохраняем самый длинный скаффолд:

````python
with open("longest.fasta", 'w', encoding='utf-8') as fout:
    fout.write(max_scaf)
````
